```python
books_and_ratings = ratings.merge(books, on='ISBN')
books_and_ratings.head()
```

```python
books_and_ratings.shape
```

```python
# Display number of ratings per book using group by function

num_ratings = books_and_ratings.groupby('Title')['rating'].count().reset_index()
num_ratings
```

```python
num_ratings.rename(columns={'rating':'number_of_ratings'},inplace=True)
num_ratings.columns 
```

```python
books_with_ratings = books_and_ratings.merge(num_ratings,on='Title')
books_with_ratings
```

```python
books_with_ratings = books_with_ratings[books_with_ratings['number_of_ratings'] >= 50]
```

```python
books_with_ratings.sample(20)
```

```python
books_with_ratings.shape
```

```python
books_with_ratings.drop_duplicates(['Title','user_id'],inplace=True)
books_with_ratings.shape
```

```python
books_with_ratings['Author'].unique()
```

We now create a Pivot Table which will generate a matrix. 

```python
book_pivot_table = books_with_ratings.pivot_table(columns='user_id',index='Title',values='rating')
book_pivot_table
```

```python
book_pivot_table.shape
```

```python
book_pivot_table.fillna(0,inplace=True)
```

```python
book_pivot_table
```

```python
from scipy.sparse import csr_matrix
```

```python
sparse_data = csr_matrix(book_pivot_table)
sparse_data
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(sparse_data)
```

```python
distance, suggestion = model.kneighbors(book_pivot_table.iloc[237,:].values.reshape(1,-1),n_neighbors=10)
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(book_pivot_table.index[suggestion[i]])
```

```python
names_of_books = book_pivot_table.index
names_of_books
```

```python
import pickle
pickle.dump(model, open('artifacts/model.pkl','wb'))
pickle.dump(names_of_books, open('artifacts/names_of_books.pkl','wb'))
pickle.dump(books_with_ratings, open('artifacts/books_with_ratings.pkl','wb'))
pickle.dump(book_pivot_table, open('artifacts/book_pivot_table.pkl','wb'))
```

Pickle files are used to create byte streams. pickle.dump() function is used to serialize a Python object and write it to a file.

We create a function to recommend books which takes book name as parameter

```python
def recommend_books(book_name):
    book_id = np.where(book_pivot_table.index == book_name)[0][0]
    #this will only return the numerical index of the book.
    distance, suggestion = model.kneighbors(book_pivot_table.iloc[book_id,:].values.reshape(1,-1),n_neighbors=10)
    for i in range(len(suggestion)):
        books = book_pivot_table.index[suggestion[i]]
        for j in books:
            print(j)
        
```

```python
book = 'Wuthering Heights'
recommend_books(book)
```
